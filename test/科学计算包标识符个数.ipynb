# numpy=600个标识符

```python
np_attr=[i for i in dir(np) if not i.startswith('_')]
np_attr
```

```python
# 学会numpy只需要掌握这些函数的用法
len(np_attr)
```

```python
type_list=[]
for i in np_attr:
    v=getattr(np,i)
    type_list.append(type(v))
from collections import Counter
pd.DataFrame(Counter(type_list).items(),columns=['name','count'])
```

numpy包含600个标识符，这些标识符分属于21个类别。

# numpy中的基本数据类型
int,numpy.bool_,float,str,tuple,bool,None,dict,list

```python
def get(type_options):
    if type(type_options)!=tuple:
        type_options=(type_options,)
    table=[]
    for i in np_attr:
        v=getattr(np,i)
        if type(v) in type_options:
            table.append((i,v,type(v)))
    return pd.DataFrame(table,columns=('name','value','type'))
get((int,np.bool_,float,str,tuple,bool,None,dict,list))
```
